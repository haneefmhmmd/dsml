```python
# Take an input from the user for start and end values. Print all even numbers between them.

start = int(input())
end = int(input())

loop_count = 0
while start <= end:
  loop_count += 1
  if start % 2 == 0: # check for even numbers
    print(start)
  start += 1

print("My loop was run", loop_count, "times!")
```

```python
# Take an input from the user for start and end values. Print all even numbers between them.

start = int(input())
end = int(input())

if start % 2 != 0: # check for odd numbers
  start += 1

loop_count = 0
while start <= end:
  loop_count += 1
  if start % 2 == 0: # check for even numbers
    print(start)
  start += 2

print("My loop was run", loop_count, "times!")
```

```python
# Take an input from the user for start and end values. Print all even numbers between them.

start = int(input())
end = int(input())

if start % 2 != 0: # check for odd numbers
  start += 1

loop_count = 0 # keep track of # of times my loop runs
while start <= end:
  loop_count += 1
  if start % 2 == 0: # check for even numbers
    print(start)
  start += 2

print("My loop was run", loop_count, "times!")
```

```python
x = 1
while x <= 100:
    if x % 3 == 0 and x % 5 == 0: # divisible by both 5 and 3 - 15, 30, 45, 60, 75, 90
        print("FizzBuzz")
    elif x % 3 == 0: # multiples of 3 - 3, 6, 9, 12, .....
        print("Fizz")
    elif x % 5 == 0: # multiples of 5 - 5, 10, 20, 25, ...
        print("Buzz")
    else:
        print(x)
    x += 1
```

```python
# Calculate sum of all numbers starting from 1 to N using a for loop.

N = int(input())
sum = 0 # initialize the sum variable with 0

for i in range(N+1): # [0, 1, 2, 3, 4, 5]
  sum += i

print(sum)
```
